```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import re
```

```python
df = pd.read_excel('akash.xlsx')
```

```python
df['Release Date'] = df['Release Date'].apply(lambda x: str(x).split(' ')[0])
```

```python
df
```

```python
def decade(x):
    if x<1930:
        return('1920\'s')
    if x<1940 and x>=1930:
        return('1930\'s')
    if x<1950 and x>=1940:
        return('1940\'s')
    if x<1960 and x>=1950:
        return('1950\'s')
    if x<1970 and x>=1960:
        return('1960\'s')
    if x<1980 and x>=1970:
        return('1970\'s')
    if x<1990 and x>=1980:
        return('1980\'s')
    if x<2000 and x>=1990:
        return('1990\'s')
    if x<2010 and x>=2000:
        return('2000\'s')
    if x<2020 and x>=2010:
        return('2010\'s')
df['Decade'] = df['Release Year'].apply(decade)
```

```python
df
```

```python
sns.set(palette='bright')
plt.figure(figsize=[15,80])
df.groupby('Director').count()
sns.barplot(y=df.groupby('Director').count().index, x='Movie Name', data=df.groupby('Director').count(), orient='h', palette='bright')
```

```python
sns.countplot(x='Decade', data=df)
```
